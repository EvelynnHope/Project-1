```python
# Import necessary libraries and dependencies
import pandas as pd
from pathlib import Path
import seaborn as sns

%matplotlib inline
```

```python
 # Set the file paths for the CSVs using the Path class from the pathlib library
Combined_portfolio =  Path("combined_close_data_SP500_group_portfolio.csv")
    

    
# Read the daily closing prices of MSFT, set the `date` as a datetime index
combined_portfolio = pd.read_csv(Combined_portfolio, index_col=0, parse_dates=True, infer_datetime_format=True)

combined_portfolio
```

```python
daily_returns = combined_portfolio.pct_change()
daily_returns.head()

daily_returns.dropna(inplace=True)

daily_returns
```

```python
for col in daily_returns.columns: 
    print(col)
```

```python
#Calculating covariance of the raw stock portfolio of daily returns

ORCL_covariance = daily_returns['ORCL'].cov(daily_returns['SP500'])
SQ_covariance = daily_returns['SQ'].cov(daily_returns['SP500'])
NVDA_covariance = daily_returns['NVDA'].cov(daily_returns['SP500'])
BABA_covariance = daily_returns['BABA'].cov(daily_returns['SP500'])
PFE_covariance = daily_returns['PFE'].cov(daily_returns['SP500'])
NFLX_covariance = daily_returns['NFLX'].cov(daily_returns['SP500'])
CRM_covariance = daily_returns['CRM'].cov(daily_returns['SP500'])
GD_covariance = daily_returns['GD'].cov(daily_returns['SP500'])
AMT_covariance = daily_returns['AMT'].cov(daily_returns['SP500'])
HD_covariance = daily_returns['HD'].cov(daily_returns['SP500'])
FB_covariance = daily_returns['FB'].cov(daily_returns['SP500'])
MSFT_covariance = daily_returns['MSFT'].cov(daily_returns['SP500'])
WMT_covariance = daily_returns['WMT'].cov(daily_returns['SP500'])
VZ_covariance = daily_returns['VZ'].cov(daily_returns['SP500'])
MCD_covariance = daily_returns['MCD'].cov(daily_returns['SP500'])
SBUX_covariance = daily_returns['SBUX'].cov(daily_returns['SP500'])
CLX_covariance = daily_returns['CLX'].cov(daily_returns['SP500'])
COST_covariance = daily_returns['COST'].cov(daily_returns['SP500'])
PSX_covariance = daily_returns['PSX'].cov(daily_returns['SP500'])
BLK_covariance = daily_returns['BLK'].cov(daily_returns['SP500'])
TROW_covariance = daily_returns['TROW'].cov(daily_returns['SP500'])
JNJ_covariance = daily_returns['JNJ'].cov(daily_returns['SP500'])
UNH_covariance = daily_returns['UNH'].cov(daily_returns['SP500'])
MMM_covariance = daily_returns['MMM'].cov(daily_returns['SP500'])
UPS_covariance = daily_returns['UPS'].cov(daily_returns['SP500'])
ACN_covariance = daily_returns['ACN'].cov(daily_returns['SP500'])
PPG_covariance = daily_returns['PPG'].cov(daily_returns['SP500'])
NEE_covariance = daily_returns['NEE'].cov(daily_returns['SP500'])
```

```python
variance = daily_returns['SP500'].var()
variance
```

```python
#Calculating covariance of stocks used to determine which stocks will stay in our portfolio and which will be dropped


ORCL_beta = ORCL_covariance / variance
SQ_beta = SQ_covariance / variance
BABA_beta = BABA_covariance / variance
PFE_beta = PFE_covariance / variance
NFLX_beta = NFLX_covariance / variance
CRM_beta = CRM_covariance / variance
GD_beta = GD_covariance / variance
AMT_beta = AMT_covariance / variance
HD_beta = HD_covariance / variance
FB_beta = FB_covariance / variance
MSFT_beta = MSFT_covariance / variance
WMT_beta = WMT_covariance / variance
VZ_beta = VZ_covariance / variance
MCD_beta = MCD_covariance / variance
SBUX_beta = SBUX_covariance / variance
CLX_beta = CLX_covariance / variance
COST_beta = COST_covariance / variance
PSX_beta = PSX_covariance / variance
BLK_beta = BLK_covariance / variance
TROW_beta = TROW_covariance / variance
JNJ_beta = JNJ_covariance / variance
UNH_beta = UNH_covariance / variance
MMM_beta = MMM_covariance / variance
UPS_beta = UPS_covariance / variance
ACN_beta = ACN_covariance / variance
PPG_beta = PPG_covariance / variance
NEE_beta = NEE_covariance / variance
```

```python
beta = {
  "ORCL": ORCL_beta,
  "SQ" : SQ_beta,
  "NVDA": ORCL_beta,
  "BABA": BABA_beta,
  "PFE": PFE_beta,
  "NFLX": NFLX_beta,
  "CRM": CRM_beta,
  "GD": GD_beta,
  "AMT": AMT_beta,
  "HD": HD_beta,
  "FB": FB_beta,
  "MSFT": MSFT_beta,
  "WMT": WMT_beta,
  "VZ": VZ_beta,
  "MCD": MCD_beta,
  "SBUX": SBUX_beta,
  "CLX": CLX_beta,
  "COST": COST_beta,
  "PSX": PSX_beta,
  "BLK": BLK_beta,
  "TROW": TROW_beta,
  "JNJ": JNJ_beta,
  "UNH": UNH_beta,
  "MMM": MMM_beta,
  "UPS": UPS_beta,
  "ACN": ACN_beta,
  "PPG": PPG_beta,
  "NEE": NEE_beta
}

beta
```

```python
sorted_beta = sorted(beta.items(),key=lambda x: x[1], reverse=False)

for i in sorted_beta: print(i[0], i[1])
```
