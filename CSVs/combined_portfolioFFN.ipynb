```python
# Import necessary libraries and dependencies
import pandas as pd
from pathlib import Path
import ffn
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
 # Set the file paths for the CSVs using the Path class from the pathlib library
Combined_portfolio =  Path("combined_close_data_SP500_group_portfolio.csv")


# Read the daily closing prices of Combined Portfolio, set the `date` as a datetime index
combined_portfolio_df = pd.read_csv(Combined_portfolio, index_col=0, parse_dates=True, infer_datetime_format=True)

combined_portfolio_df.head()
```

```python
#Assiging performance object and calculating stats from combined portfolio
perf = combined_portfolio_df.calc_stats()
```

```python
#Assigning returns for portfolio
returns = combined_portfolio_df.to_log_returns().dropna()
returns.head()
```

```python
#Histograms for returns for Blk Porfolio and SP500
ax = returns.hist(figsize=(30,15))
```

```python
#Different layout for Returns correlation
returns.corr().as_format('.2f')
```

```python
returns.plot_corr_heatmap(figsize=(30,15))
```

```python
ax = combined_portfolio_df.rebase().plot(figsize=(30,10))
```

```python
returns.calc_mean_var_weights().as_format('.2%')
```

```python
#Calculates the return / risk ratio. Basically the Sharpe ratio without factoring in the risk-free rate.
ffn.core.calc_risk_return_ratio(returns)
```

```python
#Calculates Sharpe ratio
ffn.core.calc_sharpe(returns, rf=0.0, nperiods=None, annualize=True)
```

```python
#Calculates Sortino Ratio
ffn.core.calc_sortino_ratio(returns, rf=0, nperiods=None, annualize=True)
```

```python
#Prints Returns tables for Monthly
ffn.core.deannualize(returns,12 )
```
